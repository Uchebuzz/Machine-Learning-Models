```python
############### Write your code in this cell (If applicable) ##################

import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

## For portugal

```python
#Import the dataset
basket_portugal = pd.read_csv('C:/Users/Uche Buzz/Desktop/School stuff/Data Mining and Infomatics/Dataset/Portugal_online_retail.csv')
```

```python
# View the dataset
basket_portugal.head()
```

```python
#Drop the invoice column
basket_portugal = basket_portugal.drop('InvoiceNo',axis=1)
```

```python
# Finding the frequent itemsets
# Call the apriori() function
# min_support = is to determine the support level, we can adjust this (1 - 0)

freq_items1 = apriori(basket_portugal, min_support = 0.04, use_colnames = True)
```

```python
#sorting by the range of the support
freq_items1.sort_values(by=['support'],ascending=False)
```

```python
#rules = association_rules(freq_items1, metric ="support", min_threshold = 0.1)
rules1 = association_rules(freq_items1, metric ="confidence", min_threshold = 0.03)
```

```python
rules1.sort_values(by=['support'],ascending=False)
```

----------------------------------------------------------    **Explanation** -------------------------------------------------------

I experimented with [support] and [confidence]. To see which will produce a good result 

## Important things to note

- In the first row the items **JUMBO BAG RED RETROSPOT, LUNCH BAG CARS BLUE** and **JUMBO SHOPPER VINTAGE RED PAISLEY** has a confidence level of 100%. Ths is to say if a customer purchases **JUMBO BAG RED RETROSPOT, LUNCH BAG CARS BLUE**, there is a 100% chance they will purchase **JUMBO SHOPPER VINTAGE RED PAISLEY**

- Notice in line **row 7** the item **JUMBO SHOPPER VINTAGE RED PAISLEY** is in the antecedent and the item **JUMBO BAG PINK VINTAGE PAISLEY** is in the conse with a confidence level of about 81% This is to say that there is an 81% chance that any customer that **initially** buys **JUMBO SHOPPER VINTAGE RED PAISLEY** will go on to buy **JUMBO BAG PINK VINTAGE PAISLEY**

- There is a 100% confidence level that if a customer **first** buys **JUMBO BAG PINK VINTAGE PAISLEY** they will most likely purchase **JUMBO SHOPPER VINTAGE RED PAISLEY**

```python
rules2 = association_rules(freq_items1, metric ="confidence", min_threshold = 1)
```

```python
rules2.sort_values(by=['confidence'],ascending=False)
```
